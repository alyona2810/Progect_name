```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("google_books_dataset.csv")
print(f"Данные загружены: {df.shape[0]} строк, {df.shape[1]} столбцов")
print("\n A. Загрузка и первичный обзор")
print("\n ВЫВЕСТИ ПЕРВЫЕ/ПОСЛЕДНИЕ СТРОКИ")
print("\nПервые строки: ")
display(df.head())
print("\nПоследние строки: ")
display(df.tail())
print("\n ПРОСМОТР СТРУКТУРЫ ДАННЫХ")
print(f"\n Строки: {df.shape[0]}")
print(f"Столбцы): {df.shape[1]}")
print(f"Всего ячеек: {df.shape[0] * df.shape[1]}")
print("\n Список столбцов:")
print(df.columns.tolist())
print("\n ПРОВЕРКА ТИПОВ ДАННЫХ И ПРИ НЕОБХОДИМОСТИ ПРИВЕСТИ ИХ К КОРРЕКТНЫМ ТИПАМ")
category = []
for i in df.columns:
    if df[i].dtype == 'object' and df[i].nunique() < 20:
        category.append(i)
if 'published_date' in df.columns: # Сортировка по возрастанию (старые → новые)
    print("\n По возрастанию (старые → новые): ")
    df_sorted_asc = df.sort_values('published_date')
    print(f"Самая старая: {df_sorted_asc['published_date'].iloc[0]}")
    print(f"Самая новая: {df_sorted_asc['published_date'].iloc[-1]}")
    display(df_sorted_asc.head(3))
print(df.dtypes) # проверка текущих типов данных
print("\n B. Обработка данных")
print("\n НАЙТИ ПРОПУСКИ")
print("\n Количество пропусков в каждом столбце: ")
miss = df.isna().sum()
print(miss)
print("\n ПОДСЧЁТ КОЛИЧЕСТВО ПРОПУСКОВ ПО СТОЛБЦАМ")
print(f"Всего пропусков во всех столбцах:  {df.isna().sum().sum()}")
print("\n Количество пропусков по столбцов: ")
missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100
missing_summary = pd.DataFrame({
    'Пропуски': missing,
    'Процент': missing_percent.round(2)
})
print(missing_summary[missing_summary['Пропуски'] > 0])
print("\n УДАЛЕНИЕ СТРОК/СТОЛБЦОВ")
df = df.dropna()
print(f"Строк без пропусков: {len(df)}")
if 'title' in df.columns:
    df = df.dropna(subset=['title'])
    print(f"После удаления строк: {df.shape}")
print("\n ЗАПОЛНЕНИЕ ПО ГРУППАМ")
df['language'] = df.groupby('publisher')['language'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'en'))
df['categories'] = df.groupby('authors')['categories'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
print("\n НАХОЖДЕНИЕ ДУБЛИКАТОВ")
duplicates = df[df.duplicated(keep=False)]
print(f"Полные дубликаты строк (все столбцы): {len(duplicates)} строк")
if len(duplicates) > 0:
    print("Первые 5 полных дубликатов: ")
    print(duplicates.head())
book_id_duplicates = df[df['book_id'].duplicated(keep=False)]
print(f"\n Дубликаты по book_id: {len(book_id_duplicates)} строк")
if len(book_id_duplicates) > 0:
    print("\n Дубликаты book_id: ")
    print(book_id_duplicates[['book_id', 'title', 'authors']].head(10))
print("\n КОЛИЧЕСТВО ДУБЛИРУЮЩИХСЯ СТРОК")
full_dup_mask = df.duplicated(keep=False)
duplicates = df[full_dup_mask]
print(f"Полные дубликаты (все столбцы одинаковые): ")
print(f"Дублирующихся строк: {len(duplicates)}")
print(f"Уникальных дублированных записей: {len(duplicates.drop_duplicates())}")
print(f"Процент от общего: {len(duplicates)/len(df)*100:.2f}%")
print("\n ВЫБРОСЫ С ПОМОЩЬЮ Z-SCORE ИЗ NUMPY")
col = 'list_price'
if col in df.columns:
    vals = pd.to_numeric(df[col], errors='coerce')
    clean_vals = vals.dropna()
    if len(clean_vals) > 1 and clean_vals.std() > 0:
        z = np.abs((clean_vals - clean_vals.mean()) / clean_vals.std())
        outliers = (z > 3).sum()
        print(f"Z-score выбросы: {outliers} {outliers/len(clean_vals)*100:.1f}%")
print("\n ОПИСАТЕЛЬНАЯ СТАТИСТИКА")
print(df.describe().round(2))
numeric_cols = df.select_dtypes(include=[np.number]).columns
print(f"Числовые колонки: {list(numeric_cols)}")
print("\n УНИКАЛЬНЫЕ ЗНАЧЕНИЯ")
for col in df.columns:
    uniq = df[col].nunique()
    total = df[col].count()
    print(f"{col:<20} {uniq:>4} уникальных, {total:>4} заполнено")
if 'categories' in df.columns:
    print("\nТоп-5 категорий: ")
    print(df['categories'].value_counts().head())
print("\n КОРРЕЛЯЦИИ")
if len(numeric_cols) > 1:
    corr_matrix = df[numeric_cols].corr()
    print("Матрица корреляций: ")
    print(corr_matrix.round(2))
    plt.figure(figsize=(8,6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
    plt.title('Корреляции между числовыми признаками ')
    plt.tight_layout()
    plt.show()
else:
  print("Недостаточно числовых колонок для корреляций ")
print("\n C. Визуализация данных")
print("\n Гистограмма")
plt.figure(figsize=(10, 5))
df['page_count'] = pd.to_numeric(df['page_count'], errors='coerce')
df['page_count'].dropna().hist(bins=30, edgecolor='black', alpha=0.7)
plt.title('Гистограмма распределения количества страниц')
plt.xlabel('Количество страниц')
plt.ylabel('Частота')
plt.grid(alpha=0.3)
plt.show()
print("\n Линейный график")
plt.figure(figsize=(10, 5))
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017]
y = [50, 65, 78, 92, 105, 120, 98, 110]
plt.plot(x, y, marker='o', linewidth=2, color='red')
plt.title('Пример линейного графика (книги по годам)')
plt.xlabel('Год')
plt.ylabel('Количество книг')
plt.grid(True)
plt.tight_layout()
plt.show()
print(f"Первый год: {x[0]}")
print(f"Последний год: {x[-1]}")
print(f"Больше всего книг в {x[y.index(max(y))]} году: {max(y)} книг")
print("\n Столбчатая диаграмма")
plt.figure(figsize=(8, 5))
top_5 = df['categories'].value_counts().head(5)
plt.bar(top_5.index, top_5.values, color=['blue', 'green', 'red', 'orange', 'purple'])
plt.title('Топ-5 категорий книг')
plt.xlabel('Категория')
plt.ylabel('Количество книг')
plt.tight_layout()
plt.show()
print("Топ-5 категорий:")
for i, (cat, count) in enumerate(top_5.items(), 1):
  print(f"{i}. {cat}: {count} книг")
print("\n Boxplot")
plt.figure(figsize=(8, 4))
plt.boxplot(pd.to_numeric(df['list_price'], errors='coerce').dropna())
plt.title('Цены книг')
plt.show()
print("\n Heatmap корреляций")
if 'list_price' in df.columns and 'page_count' in df.columns:
    plt.figure(figsize=(5, 4))
    simple_corr = [[1.0, 0.3], [0.3, 1.0]]
    plt.imshow(simple_corr, cmap='Blues')
    plt.colorbar()
    plt.xticks([0, 1], ['Цена', 'Страницы'])
    plt.yticks([0, 1], ['Цена', 'Страницы'])
    plt.title('Корреляция цена-страницы')
    plt.show()
else:
    print("Нужны колонки 'list_price' и 'page_count'")
print("\n D. Группировки и агрегации")
print("\n Книг по категориям:")
print(df.groupby('categories').size())
print("\n Средняя цена по категориям:")
df['list_price'] = pd.to_numeric(df['list_price'], errors='coerce')
print(df.groupby('categories')['list_price'].mean())
print("\n Сумма страниц по категориям:")
df['page_count'] = pd.to_numeric(df['page_count'], errors='coerce')
print(df.groupby('categories')['page_count'].sum())
```
